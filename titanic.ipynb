### Criar um DataFrame
Quando efetuamos a leitura de um arquivo csv, o pandas já transforma a estrutura em um dataframe. Se possuir uma dicionário e a partir dele, desejar criar um dataframe, basta chamar .DataFrame(data = variável ou estrutura de dados).

```python
dados = {
    'paises':['Brasil', 'Canadá', 'Irlanda', 'Alemanha', 'Inglaterra'],
    'habitantes': [209900000.00, 3759000.00, 4830000.00 ,82790000.00, 55980000.00]
}
```

```python
# Atribuindo a uma variável um novo DataFrame
dados_df = pd.DataFrame(data=dados)
```

```python
# visualizar resultado
dados_df
```

```python
# Selecionando as linhas entre os índices de 1 a 10
titanic_df[1:10]
```

 ### Selecionar colunas de um DataFrame
 Na seleção de colunas de um DataFrame, se passar o nome do dataframe e sua coluna:
     
   **titanic_df['Name']** 
 
 Irá retornar uma estrutura chamada Series. 
 
 Para retorna como um DataFrame, devemos fazer da seguinte forma:  
 
  **titanic_df[['Name']]** 
  
  Perceba que a diferença está no número de colchetes.
  #####  [ ] : series
  ##### [[ ]] : dataframe 

```python
# Retornando Series
titanic_df['Name']
```

```python
# Retornando coluna 'Name'
titanic_df[['Name']]
```

```python
# Selecionando colunas Name e Age
titanic_df[['Name', 'Age']]
```

Para selecionar um DataFrame por índices, podemos utilizar iloc.

.iloc[linhas, colunas]

**Exemplo:**

dataframe.iloc[:10, 0:3]

Neste caso, estamos selecionando os 10 primeiras linhas, e as colunas dos índices 0 ao 3.

```python
# utilizando iloc, para seleção de linhas e colunas
titanic_df.iloc[:5, 0:2]
```

```python
# atribuindo colunas a uma variável
titanic_df_name_age = titanic_df[['Name', 'Age']]
```

### Operações básicas com DataFrames


```python
# Analise geral dos dados do DataFrame
titanic_df.describe()
```

```python
# verificando se o DataFrame possui o nome Braund, Mr. Owen Harris
por_nome = titanic_df_name_age['Name'] == 'Braund, Mr. Owen Harris'
```

```python
# retornando Dataframe completo, com a condição
titanic_df_name_age[por_nome]
```

```python
# verificando se o DataFrame possui idades menores do que 18
# note que está sendo selecionado apenas os campos que não são nulos
menor_de_idade = titanic_df_name_age['Age'].notnull() <= 18
```

```python
# retornando dados dos menores de idade
titanic_df_name_age[menor_de_idade]
```

```python
# identificando campos que estão nulos na coluna Age
idades_nulas = titanic_df_name_age['Age'].isnull()
```

```python
# retornando as linhas que a idade está nula
titanic_df_name_age[idades_nulas]
```

```python
# verificando pessoas menores de idade e que não sobreviveram
menores_nao_sobreviventes = (titanic_df['Age'] <= 18) & (titanic_df['Survived'] == False)
```

```python
# atribuindo dados de seleção em variável
titanic_df_menores_nao_sobreviventes = titanic_df[menores_nao_sobreviventes]
```

```python
titanic_df_menores_nao_sobreviventes
```

```python
# retornando contagem do número de linhas
titanic_df_menores_nao_sobreviventes.count()
```

```python
# visualizando contagem no dataframe com todos os dados
titanic_df.count()
```

Funções Lambda em python, é uma maneira mais rápida de criar uma única expressão.
para saber mais, [clique aqui](https://www.w3schools.com/python/python_lambda.asp).

Abaíxo estamos adicionando um novo campo no DataFrame. Para saber mais sobre, [clique aqui](https://datatofish.com/if-condition-in-pandas-dataframe/).

### Criar nova coluna
Para criar uma nova coluna basta chamar o dataframe['nome da coluna']

```python
# Criando um novo campo booleano chamado eh_menor
titanic_df['eh_menor'] = titanic_df['Age'].apply(lambda x: True if x < 18 else False)
```

```python
# importar bibliotecas
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# aplicando estilo no seaborn
sns.set(style="darkgrid")
```

```python
# criando gráfico comparando o número de sobreviventes e não sobreviventes
sns.countplot(x='Survived', data=titanic_df, hue='Survived', linewidth=0)
```

```python
# gráfico de comparação de quantidade de pessoas por classe do návio
sns.countplot(y='Pclass', data=titanic_df, hue='Pclass')
```

```python
# colocando um novo estilo e criando gráfica de número de sobreviventes por classe
sns.set(style="ticks")
sns.catplot(x='Pclass', data=titanic_df, hue='Sex', col='Survived', kind='count', 
            height=6, aspect=.7)
```
